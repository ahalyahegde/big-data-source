```
from pyspark.sql.functions import lit
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.master('local') \
                     .appName('Practice Spark') \
                     .getOrCreate()
```

```
df = spark.read.options(header=True, inferSchema=True) \
               .csv('/FileStore/tables/employees-1.csv') 
               
df.printSchema()
```

```
# Select only Gender column
df.select(df.Gender).show()

# Create new column with constant value
df.withColumn('Country', lit('USA')).show()

# Rename column
df.withColumnRenamed('Gender', 'SEX').show()

# Drop column
df.drop('Country').show()

#Change data type
df.withColumn('salary', col('salary').cast("string")).printSchema()
```

```
sparkContext = spark.sparkContext
rdd =sparkContext.parallelize([1,2,3,4], 2)
print(rdd.collect())
print(rdd.getNumPartitions())
print(rdd.glom.collect())
```
